###### Describe of the data set which is called heart_failure_clinical_records_dataset.csv

Heart failure clinical records Data Set contains the medical records of 299 patients who had heart failure.
The dataset contains 11 clinical features (some of them are binary, others are numerical), the follow-up period and the label DEATH_EVENT that indicates whether or not the patient has died.
We can find some features strictly related to medical aspects like levels of enzymes, sodium, creatinine and platelets in the blood and others that are more common like age, sex or smoking.

The dataset is collected in 2015 at the Allied Hospital in Faisalabad (Punjab, Pakistan)
0. age = Age
1. anaemia = Decrease of red blood cells or hemoglobin (haematocrit levels were lower than 36%) (0 = Higher than 36%, 1 = Lower than 36%)
2. creatinine_phosphokinase = Level of the CPK enzyme in the blood
3. diabetes = If the patient has diabetes (0 = Don't have diabetes, 1 = Have Diabetes)
4. ejection_fraction = Percentage of blood leaving the heart at each contraction
5. high_blood_pressure = If a patient has hypertension (0 = Don't have High-blood Pressure, 1 = Have High-blood Pressure) 
6. platelets = Platelets in the blood
7. serum_creatinine = Level of creatinine in the blood
8. serum_sodium = Level of sodium in the blood
9. sex = Woman or man (0 = Woman, 1 = Man)
10. smoking = If the patient smokes (0 = No Smoking, 1 = Smoking) 
11. time = Follow-up period
12. DEATH_EVENT = If the patient died during the follow-up period (0 = alive, 1 = death)

The details and procedure-code of the decision tree algorithm are briefed below:

1. The function of the step of createNode() expands the size of the decision tree classifier by establishing a new node (branch) in the decision tree model. A node in the decision tree classifier either a test condition, represented as the procedure code of node.test_cond, or a class label, represented as the code of node.label.

2. The function of the find_best_split() is used to ensure which feature should be chosen as the test condition for splitting the training records from the dataset. 3. For the Classify() function identifies which class label to be distributed to the leaf node part. Each of these leaf node t, p(i|t) will refers to the fraction of training results from class i related with the node t. Recently, leaf node is setup for the class which has lots of training records: 
            leaf.label = argmax p(i|t)
   while the argmax operator will then return the argument of i which will maximizes the p(i|t) expression. On the other hand, it will offer the information for identifying the class label of the leaf node. 4. For the stopping_cond(), it is used to stop the tree-growing continuously. It is processed by checking if all the records are having the similar class label or having the similar feature values. On the other hand, it can be stopped by checking if the records have been lower than the minimum threshold. 

After modelling the whole decision tree, tree-pruning will be triggered to minimize the sizes of the decision tree.
